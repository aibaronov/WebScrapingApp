```python
import requests
from bs4 import BeautifulSoup

r = requests.get("https://www.century21.com/real-estate/los-angeles-ca/LCCALOSANGELES/")
c=r.content

soup = BeautifulSoup(c,"html.parser")

all = soup.find_all("div",{"class":"property-card-primary-info"})

all[0].find("a",{"class": "listing-price"}).text

all[0].find("a",{"class": "listing-price"}).text.replace("\n","").replace(" ","")
```

```python
l = []
for item in all:
    d = {}
    d["Price"] = item.find("a",{"class","listing-price"}).text.replace("\n","").replace(" ","")
    d["Address"] = item.find_all("div",{"class":"property-address-info"})[0].text.replace("\n","")
    try:
        d["Beds"] = item.find("div",{"class","property-beds"}).find("strong").text
    except:
        d["Beds"] = None
    try:
        d["Baths"] = item.find("div",{"class","property-baths"}).find("strong").text
    except:
        d["Baths"] = None
    try:
        d["Square feet"] = item.find("div",{"class","property-sqft"}).find("strong").text
    except:
        d["Square feet"] = None
    l.append(d)
```

```python
import pandas
df = pandas.DataFrame(l)
```
